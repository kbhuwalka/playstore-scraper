```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10.0, 10.0)
import pandas as pd
from mpltools import style
from matplotlib import font_manager as fm
style.use("ggplot")
appListTimestamp = "201401221412"
```

```python
#All apps with permissions
totalDF = pd.read_csv("../%s/data/permission/total_permission.csv"%appListTimestamp, sep = " , ", index_col=False)
```

```python
appWithPermission = totalDF['count'][0]
totalAppCount = 918597 #for 201401221412
```

```python
#Draw pie figure show the percentage of apps with permissions

patches, texts, autotexts = plt.pie([appWithPermission, totalAppCount - appWithPermission], colors = ['lightseagreen','0.75'], explode = (0, 0.05), labels = ('Apps with permission','Apps without permission'), autopct='%1.1f%%', shadow=True,
labeldistance=0.3)
#for set font size
proptease = fm.FontProperties()
proptease.set_size('xx-large')
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)

savefig('../%s/figures/appWithPermission.svg'%appListTimestamp)
```

```python
#method which draw permission rank list
def drawPermission(fileName):
    print fileName
    ctgryDF = pd.read_csv(fileName, sep = " , ", index_col=False)
    ctgryDF[:10]
    total = float(ctgryDF['count'][0])
    ctgryDF['count'] = ctgryDF['count']/total*100
    ctgryDF = ctgryDF[1:]
    
    fig, ax = subplots()
    plt.barh(arange(0.6,20,1), ctgryDF['count'][:20])
    yt = yticks(range(1,21),[entry for entry in ctgryDF['permission'][:20]])
    loc =  matplotlib.ticker.MultipleLocator(base=5.0)
    ax.xaxis.set_major_locator(loc)
    plt.ylim((0,21))
    plt.gca().invert_yaxis()
    xlabel = ax.set_xlabel("Popularity (percentage of apps)")
    ax.xaxis.set_label_position('top') 
    ax.xaxis.tick_top()
    plt.suptitle("Top 20 most used permissions in %s category (Play Store Jan. 2014)"%(fileName.split('/')[-1][:-15].replace("_"," ")), verticalalignment = "bottom")
    savefig(fileName.split('data')[0]+'figures/mostUsedPermission_%s.svg'%(fileName.split('/')[-1][:-15]), bbox_extra_artists=[xlabel], bbox_inches='tight')
```

```python
import os
for fileName in os.listdir("../%s/data/permission/"%appListTimestamp):
    drawPermission("../%s/data/permission/%s"%(appListTimestamp,fileName))
```
