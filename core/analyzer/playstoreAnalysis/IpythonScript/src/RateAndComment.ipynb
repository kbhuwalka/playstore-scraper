```python
#ratingS is the starRating whose ratingCount no less than 100
ratingS = fec[fec['ratingsCount']>=100]['starRating']
```

```python
#starRating hisogram; ratingCount no less than 100
fig, ax = subplots()
ax.hist(ratingS, bins=200)
ax.set_xlabel("Star Rating")
ax.set_ylabel("Number of Apps")
ax.set_title("App Star Rating Histogram (for apps with >= 100 rating, Play Store %s)"%timeStr)
savefig("../%s/figures/appStarRatingHist.svg"%appListTimestamp)
```

```python
ratingSumS = fec['ratingsCount']*fec['starRating']
```

```python
ratingSumS = ratingSumS.order(ascending=False)
ratingSumS[:10]
```

```python
#Top 10 rating apps by total rating
import matplotlib.ticker as plticker
packageNames = ratingSumS.index
ratingSumS.index=range(1,len(ratingSumS)+1)
fig, ax = subplots()
ax.plot( ratingSumS.index[:10], ratingSumS[:10], "o--")
ax.set_xlabel("Ranking (1-10)")
ax.set_ylabel("Total Rating (star rating * number of rating)")
ax.set_title("Top 10 rating apps by total rating(Play Store %s)"%timeStr)
loc = plticker.MultipleLocator(base=1.0)
ax.xaxis.set_major_locator(loc)
for index in range(10):
    ax.annotate(packageNames[index], xy=(index+1, ratingSumS[index+1]), xytext = (pow(-1, index)*60, pow(-1, index)*20),
            textcoords = 'offset points', arrowprops=dict(facecolor='lightseagreen', shrink=0.01),)
plt.xlim((0,11))
plt.ylim((0.1e7,8.5e7))
savefig("../%s/figures/top10RatingAppBySum.svg"%appListTimestamp)
```

```python
ratingCountS = fec['ratingsCount'].order(ascending=False)
ratingCountS[:10]
```

```python
#Top rating app by rating count
packageNames = ratingCountS.index
ratingCountS.index = range(len(ratingCountS))

ratingCountS.index=range(1,len(ratingCountS)+1)
fig, ax = subplots()
ax.plot( ratingCountS.index[:10], ratingCountS[:10], "o--")
ax.set_xlabel("Ranking (1-10)")
ax.set_ylabel("Total Number of Rating")
ax.set_title("Top 10 rating apps by total number of ratings (Play Store %s)"%timeStr)
loc = plticker.MultipleLocator(base=1.0)
ax.xaxis.set_major_locator(loc)
for index in range(10):
    ax.annotate(packageNames[index], xy=(index+1, ratingCountS[index+1]), xytext = (pow(-1, index)*60, pow(-1, index)*20),
            textcoords = 'offset points', arrowprops=dict(facecolor='lightseagreen', shrink=0.01),)
plt.xlim((0,11))
plt.ylim((0.1e7,2.3e7))
savefig("../%s/figures/top10RatingAppByCount.svg"%appListTimestamp)
```
